54) ¿Cuál es la acción más realizada por usuarios que no están registrados? (⭐⭐)

```python
df_logs = pd.read_csv('/content/drive/MyDrive/Wikipedia dataset 2c2021/logs.csv')
df_logs.head(20)
```

filtro el dataframe por los usuarios sin id ya que son los que no tienen usuario

```python
df_nr = df_logs.loc[df_logs['contributor_id'].isnull()]
```

Aparece un contributor_username 'Unknown user' pero es similar a NaN, no es un usuario real.

```python
df_nr
```

Agrupo por action, para contar la cantidad de veces que fue realizada cada una

```python
df_mas_realizados = df_nr.groupby(['action']).agg({'item_id':'count'})
```

```python
df_mas_realizados.columns = ['veces_realizada']
df_mas_realizados
```

```python
maxima_cantidad = df_mas_realizados['veces_realizada'].max()
df_mas_realizada = df_mas_realizados.loc[df_mas_realizados['veces_realizada'] >= maxima_cantidad]
df_mas_realizada
```

Por lo que podemos ver que el máximo es create con 124728
